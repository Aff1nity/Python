```python
# Some observations:
# The mean value of the 'popularity' column is 27.57, which indicates that the average popularity score is moderate.
# The standard deviation of the 'popularity' column is 18.37, which suggests that the popularity scores are somewhat spread out around the mean.
# The minimum popularity score is 0, which means there are some tracks with very low popularity.
# The 25th percentile of the popularity score is 13, which indicates that 25% of the tracks have a popularity score of 13 or lower.
# The median popularity score is 27, which means that 50% of the tracks have a popularity score of 27 or lower.
# The 75th percentile of the popularity score is 41, which indicates that 75% of the tracks have a popularity score of 41 or lower.
```

```python
# Question: Is popularity highly correlated to danceability?
```

```python
df_tracks.plot(kind='scatter', x='popularity', y='danceability', figsize=(15, 10))

# set the plot title and axis labels
plt.title('Popularity vs. Danceability')
plt.xlabel('Popularity')
plt.ylabel('Danceability')

# display the plot
plt.show()
```
